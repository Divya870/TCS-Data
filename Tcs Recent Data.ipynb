```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
tcs=yf.download('TCS.NS',
  start='2020-01-01',
  end='2025-02-25',
  progress=False)
```

```python
tcs['simple_rtn']=tcs.Close.pct_change()
tcs['simple_rtn']
```

```python
tcs['log_rtn']=np.log(tcs.Close/tcs.Close.shift(1))
tcs['log_rtn']
```

```python
tcs['vsimple_rtn']=tcs.Volume.pct_change()
tcs['vsimple_rtn']
```
